```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data  = pd.read_csv("/features_3_sec.csv")
data.head(50)
```

```python
data.tail()
```

```python
def calculate_centroid_and_spread(class_data):
    centroid = np.mean(class_data)
    spread = np.std(class_data)
    return centroid, spread

def calculate_distance(centroid1, centroid2):
    distance = np.linalg.norm(centroid1 - centroid2)
    return distance

def print_results(centroid1, centroid2, spread1, spread2, distance):
    print("Mean (centroid) of class 1:", centroid1)
    print("Mean (centroid) of class 2:", centroid2)
    print("Spread (standard deviation) of class 1:", spread1)
    print("Spread (standard deviation) of class 2:", spread2)
    print("Distance between centroids of class 1 and class 2:", distance)

# Assuming you have two classes represented by their feature vectors
class1_data = data['rms_mean']
class2_data = data['rms_var']

# Calculate centroid and spread for each class
centroid_class1, spread_class1 = calculate_centroid_and_spread(class1_data)
centroid_class2, spread_class2 = calculate_centroid_and_spread(class2_data)

# Calculate distance between centroids
distance_between_centroids = calculate_distance(centroid_class1, centroid_class2)

# Print results
print_results(centroid_class1, centroid_class2, spread_class1, spread_class2, distance_between_centroids)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_centroid_and_spread(class_data):
    centroid = np.mean(class_data)
    spread = np.std(class_data)
    return centroid, spread

def calculate_distance(centroid1, centroid2):
    distance = np.abs(centroid1 - centroid2)
    return distance

def plot_data_with_centroids(class1_data, class2_data, centroid_class1, centroid_class2, spread_class1, spread_class2, distance_between_centroids):
    plt.scatter(range(len(class1_data)), class1_data, label='Class 1 Data')
    plt.scatter(range(len(class2_data)), class2_data, label='Class 2 Data')

    plt.errorbar(centroid_class1, centroid_class1, yerr=spread_class1, fmt='r*', label='Class 1 Centroid')
    plt.errorbar(centroid_class2, centroid_class2, yerr=spread_class2, fmt='r*', label='Class 2 Centroid')

    plt.text((centroid_class1 + centroid_class2) / 2, min(centroid_class1, centroid_class2), f'Distance: {distance_between_centroids:.2f}', ha='center', color='blue')

    plt.xlabel('Samples')
    plt.ylabel('Feature Values')
    plt.title('Dataset with Centroids and Spread')
    plt.legend()
    plt.grid(True)
    plt.show()

# Assuming 'class1_data' and 'class2_data' are NumPy arrays containing the data
class1_data = np.array(class1_data)
class2_data = np.array(class2_data)

# Calculate centroids and spreads for each class
centroid_class1, spread_class1 = calculate_centroid_and_spread(class1_data)
centroid_class2, spread_class2 = calculate_centroid_and_spread(class2_data)

# Calculate distance between centroids
distance_between_centroids = calculate_distance(centroid_class1, centroid_class2)

# Plot data with centroids and spread
plot_data_with_centroids(class1_data, class2_data, centroid_class1, centroid_class2, spread_class1, spread_class2, distance_between_centroids)
```
