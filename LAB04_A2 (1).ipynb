```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("/content/features_30_sec.csv")
```

```python
def calculate_mean_and_variance(feature_data):
    mean = np.mean(feature_data)
    variance = np.var(feature_data)
    return mean, variance

def plot_histogram(feature_data, feature_column):
    plt.hist(feature_data, bins=10)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Histogram of {}'.format(feature_column))
    plt.show()

# Assuming df is a DataFrame and feature_column is the column name
feature_column = 'spectral_centroid_mean'

# Extract the feature data
feature_data = df[feature_column]

# Calculate mean and variance
mean, variance = calculate_mean_and_variance(feature_data)

# Plot histogram
plot_histogram(feature_data, feature_column)

print("Mean:", mean)
print("Variance:", variance)
```
